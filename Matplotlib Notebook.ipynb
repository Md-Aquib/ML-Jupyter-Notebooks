```python
import pandas as pd
cars_sales = pd.read_csv(r'car-sales.csv')
```

```python
cars_sales.head()
```

```python
cars_sales['Price'] = cars_sales['Price'].str.replace('[\$\,\.]','')
```

```python
cars_sales['Price'] = cars_sales['Price'].str[:-2]
```

```python
cars_sales.head()
```

```python
cars_sales['Price'] = cars_sales['Price'].astype(int)
```

```python
cars_sales['Price'].dtypes
```

```python
cars_sales['Total_Sales'] = cars_sales['Price'].cumsum()
```

```python
cars_sales['Date_Sold'] = pd.date_range('1/1/2022',periods=len(cars_sales))
```

```python
cars_sales
```

```python
cars_sales.plot(x='Date_Sold',y='Total_Sales')
```

```python
cars_sales.plot.scatter(x='Odometer (KM)',y='Price')
```

```python
heart_disease = pd.read_csv(r'heart-disease.csv')
```

```python
heart_disease.head(10)
```

```python
heart_disease['age'].plot.hist(bins=50)
```

```python
heart_disease.plot.hist(subplots=True)
```

```python
heart_disease.plot.hist(subplots=True, figsize=(10,30))
```

```python
age_50 = heart_disease[heart_disease['age']>50]
age_50.head()
```

```python
# creating plot
fig , ax = plt.subplots()
# plotting the data
scatter = ax.scatter(x=age_50['age'],y=age_50['chol'], c = age_50['target'])
# customizing the plot
ax.set(title='Heart Disease and Cholesterol Levels',xlabel='Age',ylabel='Cholesterol')
# Adding Legends
ax.legend(*scatter.legend_elements(),title='Target')
# Adding Mean Line
ax.axhline(age_50['chol'].mean(),linestyle='--')
```

```python
# creating plot
fig , ax = plt.subplots()
# plotting the data
scatter = ax.scatter(x=age_50['age'],y=age_50['chol'], c = age_50['target'],cmap='winter')
# customizing the plot
ax.set(title='Heart Disease and Cholesterol Levels',xlabel='Age',ylabel='Cholesterol')
# Adding Legends
ax.legend(*scatter.legend_elements(),title='Target')
# Adding Mean Line
ax.axhline(age_50['chol'].mean(),linestyle='--',color='red')
```
